```python
# Lendo imagens
ufabc = io.imread('/content/drive/My Drive/pdi-2022/ufabc.png')
ufabc = cv.resize(ufabc, (600,800))

ufabcEspelhada = cv.flip(ufabc,1)

sa = io.imread('/content/drive/My Drive/pdi-2022/campus-sa.jpg')
sa = cv.resize(sa, (600,800))

sbc = io.imread('/content/drive/My Drive/pdi-2022/campus-sbc.jpg')
sbc = cv.resize(sbc, (600,800))

logos = np.hstack([ufabc, ufabcEspelhada])
campus = np.hstack([sa,sbc])
colagem = np.vstack([logos, campus])

plt.figure(figsize = (20,10))
plt.imshow(colagem);
```
